```python
# Importing Modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading CSV DATA
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df.info()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
# Number of Real and Fraud Transaction
df["Class"].value_counts()
#HIGHLY UNBALANCED
```

```python
#Seperating both Transactions
real=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
real.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing wrt Class
df.groupby("Class").mean()
```

```python
real_sample=real.sample(n=492)
real_sample.shape
```

```python
newdf=pd.concat([fraud,real_sample],axis=0)
newdf.head()
```

```python
newdf["Class"].value_counts()
```

```python
newdf.groupby("Class").mean()
```

```python
# Splitting data into features and target
x=newdf.drop(columns="Class")
y=newdf["Class"]
```

```python
#Split datafor training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#MODEL TRAINING
model=LogisticRegression(max_iter=100000)
model.fit(x_train,y_train)
```

```python
# Evaluation
train_pred=model.predict(x_train)
accuracy_score(y_train,train_pred)
```

```python
# Accuracy on Test Inputs
test_pred=model.predict(x_test)
accuracy_score(y_test,test_pred)
```
